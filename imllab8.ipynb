```python

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/HCLTECH.csv')

# Limit the DataFrame 
limited_df = df.iloc[:100]

# Create a date range with the desired start date and frequency
date_range = pd.date_range("2000-03-07", periods=len(limited_df), freq="D")

# Set the date range as the index of the limited DataFrame
limited_df.set_index(date_range, inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(limited_df.index, limited_df['Open'], label='Opening Price', marker='o')
plt.plot(limited_df.index, limited_df['Close'], label='Closing Price', marker='o')

# Set plot labels and title
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Opening and Closing Stock Prices')

# Show a legend
plt.legend()

# Show the plot

plt.show()




```
